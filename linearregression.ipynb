```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Advertising.csv");
df.head()
```

```python
df['total_spent'] = df['TV']+df['Radio']+df['Newspaper'];
df.head()
```

```python
sns.scatterplot(data=df,x='total_spent', y='Sales')
```

```python
sns.regplot(data = df,x='total_spent',y='Sales')
# regplot = shows best fit line regression line using sea born library
```

```python
help(np.polyfit)
```

```python
X = df['total_spent'];
y = df['Sales'];
np.polyfit(X,y,deg=1)
```

```python
potential_spend = np.linspace(0,500,100)
predicted_sales = 0.04868788*potential_spend + 4.24302822;
```

```python
#sns.scatterplot(data=df,x='total_spent', y='Sales',color='red')
sns.regplot(data = df,x='total_spent',y='Sales',color='red')
plt.plot(potential_spend,predicted_sales,color='blue')
```
